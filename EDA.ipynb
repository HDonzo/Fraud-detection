```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('/mnt/c/Users/donov/Bureau/Fraude/Data/creditcard.csv', low_memory = True)
```

```python
Data.info()
```

```python
# Première visualisation

# Visualisation de premières lignes

print(Data.head())

# Affichage des informations du Dataframe

print(Data.info())

# Affichage des statistiques descriptives de chaque Dataframe

print(Data.describe(include='all'))
```

```python
# Phase de Nettoyage de Données 

# Verification des valeurs NaN ou doublons dans le Dataframe 

#Valeur NaN
print(print(Data.isnull().sum()))

# Suppression des valeurs NaN
Data.dropna(how='all',inplace=True)

# Doublons
print(f"Doublons : {Data.duplicated().sum()}")

# Suppression des doublons
Data.drop_duplicates(inplace=True)

# Ajout d'une nouvelle colonne heure_journée et jours

Data['heure_journée'] = (Data['Time'] % 86400) // 3600

Data['jours'] = (Data['Time'] // 86400) + 1
```

```python
# Seconde visualisation

# Visualisation de premières lignes

print(Data.head())

# Affichage des informations du Dataframe

print(Data.info())

# Affichage des statistiques descriptives de chaque Dataframe

print(Data.describe(include='all'))

# Sauvegarder Les Datas nettoyés pour les autres étapes

Data_Processed = Data.to_csv("Data_Processed.csv", index=False)
```

```python
# Premiere Exploration : Répartition des fraudes 1 et des non fraudes 0 : Camembert

# Comptes (assure l’ordre 0 puis 1)
counts = Data['Class'].value_counts().sort_index()
labels = ['Non fraude (0)', 'Fraude (1)']
colors = ['skyblue', 'salmon']
total = counts.sum()

plt.figure(figsize=(6, 6))

# 1) Pourcentage + effectif
wedges, texts, autotexts = plt.pie(
    counts.values,
    labels=None,                
    colors=colors,
    startangle=90,
    counterclock=False,
    explode=[0, 0.08],
    autopct=lambda p: f"{p:.2f}%\n({int(round(p/100*total)):,})",
    pctdistance=0.7
)

# 2) Ajouter du label sous les données
for i, a in enumerate(autotexts):
    a.set_text(a.get_text() + f"\n{labels[i]}")
    a.set_ha('center')

plt.title("Répartition des transactions : Non fraude (0) vs Fraude (1)")
plt.axis('equal') 
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Fraude vs non Fraude 1.png")
plt.tight_layout()
plt.show()

```

```python
# Premiere Exploration : Répartition des fraudes 1 et des non fraudes 0 : Barplot

Fraud_distrib = Data['Class'].value_counts(normalize=True).sort_index() * 100

plt.figure(figsize=(6,4))
ax = Fraud_distrib.plot(kind='bar',
                        color=['skyblue', 'salmon'],
                        alpha=0.85)

ax.set_title("Répartition des Fraudes (1) et Transactions Normales (0) [%]")
ax.set_xlabel("Type de Transaction")
ax.set_ylabel("Pourcentage (%)")
ax.set_ylim(0, max(105, Fraud_distrib.max() * 1.15))  
ax.set_xticklabels(['0', '1'], rotation=0)            # le 1 et le 0 avaient une rotation anormale

for index, value in enumerate(Fraud_distrib):
    plt.text(index, value + 1, f"{value:.3f}%", ha='center', va='bottom', fontsize=10) # % au dessus des barplots
    
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Fraude vs non Fraude 2.png")
plt.tight_layout()
plt.show()
```

```python
# Seconde exploration : Visualisation de la distribution de amount selon class : Histogramme superposé

plt.figure(figsize=(10,6))

# Histogrammes par classe
sns.histplot(Data[Data['Class']==0]['Amount'], bins=100, color='skyblue', label='Non fraude (0)', alpha=0.6, stat="density")
sns.histplot(Data[Data['Class']==1]['Amount'], bins=100, color='salmon', label='Fraude (1)', alpha=0.6, stat="density")

plt.legend()
plt.xlabel("Montant de la transaction")
plt.ylabel("Densité")
plt.title("Distribution de Amount selon Class")
plt.xlim(0, 2000)  # optionnel : zoomer pour mieux voir
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Distribution amount-class hist.png")
plt.show()
```

```python
# Seconde exploration : Visualisation de la distribution de amount selon class : Boxplot

plt.figure(figsize=(8,6))
sns.boxplot(x='Class', y='Amount', data=Data, palette=['skyblue','salmon'])
plt.title("Distribution des montants par classe (0=Non fraude, 1=Fraude)")
plt.xlabel("Classe")
plt.ylabel("Montant (€)")
plt.ylim(0, 2000)  # zoom pour éviter que les valeurs extrêmes écrasent le graphe
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Distribution amount-class boxplt.png")
plt.show()
```

```python
# Troisieme exploration : Distribution des transactions selon l'heure de la journée

plt.figure(figsize=(10,6))
sns.histplot(data=Data, x='heure_journée', hue='Class', bins=24, stat='density', 
             palette=['salmon', 'skyblue'], alpha=0.6, multiple='stack')

plt.title("Distribution des transactions selon l'heure de la journée")
plt.xlabel("Heure de la journée (0h → 24h)")
plt.ylabel("Densité")
plt.legend(labels=['Non fraude (0)','Fraude (1)'])
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Comparaison time-class hist.png")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=Data, x='Class', y='heure_journée', palette=['skyblue','salmon'])
plt.title("Comparaison de l'heure des transactions selon la classe")
plt.xlabel("Classe (0=Non fraude, 1=Fraude)")
plt.ylabel("Heure de la journée")
plt.xticks([0,1], ['Non fraude (0)', 'Fraude (1)'])
plt.savefig("/mnt/c/Users/donov/Bureau/Fraude/Data/Outputs/Comparaison time-class boxplt.png")
plt.tight_layout()
plt.show()
```
